```python
def r(a):
    b = u.ufloat(44, 0.01 / 3**0.5)
    c = u.ufloat(117.12, 0.01 / 3**0.5)
    d = u.ufloat(49.90, 0.03 / 3**0.5)
    e = u.ufloat(25.04, 0.01 / 3**0.5)
    p = u.ufloat(120, 0.01 / 3**0.5) * 0.5 - u.ufloat(17.1, 0.01 / 3**0.5) * 0.5
    
    beta = a - 0.5 * b - 0.5 * d
    gamma = c - 0.5 * d - 0.5 * e

    alpha = (gamma**2 - beta**2)**0.5

    return (beta**2 + (alpha - p)**2)**0.5
```

```python
print(r(u.ufloat(97.71, 0.01 / 3**0.5)))
print(r(u.ufloat(98.23, 0.01 / 3**0.5)))
```
