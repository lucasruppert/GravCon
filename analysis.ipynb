```python
import numpy as np
import uncertainties as u
```

```python
def r(a, b, c):
    x = 42e-3
    p = (120 - 0.5 * 17.1) * 1e-3
    r_s = u.ufloat(49.90, 0.03) * 1e-3 * 0.5

    beta = b + r_s - (a - 0.5 * x)
    gamma = 0.5 * c - r_s
    alpha = (gamma**2 - beta**2)**0.5

    return (beta**2 + (alpha - 0.5 * p)**2)**0.5
```

```python
print(
    r(
    a=u.ufloat(0.209, 0.001),
    b=u.ufloat(0.22, 0.001),
    c=u.ufloat(0.207, 0.002)
)
)
```
